```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(1,10,20)
```

```python
x
```

```python
y = x**2
```

```python
y
```

```python
#Functional
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r')
plt.xlabel('X Value')
plt.ylabel('X Value')
plt.title('X vs Y')
```

```python
plt.subplot(1,3,1)
plt.plot(x,y,'r')

plt.subplot(1,3,2)
plt.plot(y,x,'g')

plt.subplot(1,3,3)
plt.plot(y,x,'b')
```

```python
plt.figure(figsize=(12,5))
plt.plot(x,y,'r')
plt.show()
```

```python
plt.savefig('x_y_plot.png', dpi=500)
```

```python
plt.figure(figsize=(12,6))
plt.plot(x,x,'b', label='x vs x')
plt.plot(x,x**2,'r',label='x vs x**2')
plt.plot(x,x*3,'g',label='x vs x**3')
plt.legend(loc=0)
```

```python
plt.figure(figsize=(12,6))
plt.plot(x,x,'b', label='x vs x')
plt.plot(x,x**2,'r',label='x vs x**2')
plt.plot(x,x*3,'g',label='x vs x**3')
plt.legend(loc=(0.1,0.2))
```

<b>Plot Appearance:<b/>

```python
plt.figure(figsize=(10,4))
plt.plot(x,y, label='X vs Y', color='black', linewidth=6) #Use rgb hex-code for more colors
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.title('X vs Y Plot')
plt.legend()
```

```python
plt.figure(figsize=(10,4))
plt.plot(x,y, label='X vs Y', color='black', linewidth=6, alpha=0.2) #Use rgb hex-code for more colors
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.title('X vs Y Plot')
plt.legend()
```

```python
plt.figure(figsize=(10,4))
plt.plot(x,y, label='X vs Y', color='red', linewidth=6, linestyle='--') #Use rgb hex-code for more colors
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.title('X vs Y Plot')
plt.legend()
```

```python
plt.figure(figsize=(10,4))
plt.plot(x,y, label='X vs Y', color='purple', linewidth=6, linestyle=':') #Use rgb hex-code for more colors
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.title('X vs Y Plot')
plt.legend()
```

```python
plt.figure(figsize=(10,4))
plt.plot(x,y, label='X vs Y', color='orange', linewidth=2, linestyle='-', marker='o') #Use rgb hex-code for more colors
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.title('X vs Y Plot')
plt.legend()
```

```python
plt.figure(figsize=(10,4))
plt.plot(x,y, label='X vs Y', color='blue', linewidth=0.5, linestyle='-', marker='1', markersize=20) #Use rgb hex-code for more colors
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.title('X vs Y Plot')
plt.legend()
```
